```python
#Importing all the modules
import pandas as pd
import numpy as np
import nltk
import os
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
```

```python
# Importing the data into a pandas dataset
path = './Assignment-2-siddharth-purii/Data/LabeledText.xlsx'
df_columns = ['file', 'Text', 'Sentiment']
df = pd.read_excel(path, names = df_columns)
# Removing the unnecessary columns.
df = df[['Text','Sentiment']]
```

```python
# Split the dataset into testing (10%) and training(90%)
size = int(len(df) * 0.90)
train_set = df[:size]
test_set = df[size:]
```

```python
#Using Vader
sia = SentimentIntensityAnalyzer()
```

```python
# Run the polarity score on the entire dataset
for index, row in df.iterrows():
    output = sia.polarity_scores(row['Text'])
    # Keeping the highest value of the sentiments
    result = output["compound"]
    if result > 0:
        print("Predicted Sentiment: positive")
    if result < 0:
        print("Predicted Sentiment: negative")
    if result == 0:
        print("Predicted Sentiment: neutral")
    print("Actual Sentiment - " + row['Sentiment'])
    print("")
    
```
